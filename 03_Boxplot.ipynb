### Setting labels

```python
sns.boxplot(x="survived", 
            y="age", showmeans=True,meanprops={"marker":"o",
                                              "markersize": "14",
                                              "markeredgecolor":"red"},
            data=kashti
           )
plt.xlabel("How many susrvide?", size=16, weight='bold')
plt.ylabel("Age(Years)")
plt.title("Box plot of survived")
```

### Facet_grid and facet_wrap

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
tip=sns.load_dataset("tips")
# Form a facetgrid using columns with a hue
graph = sns.FacetGrid(tip, col ="sex",  hue ="day")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()
# show the object
plt.show()
```

```python
# Form a facetgrid using columns with a hue
graph = sns.FacetGrid(tip, row ='smoker', col ='time')
# map the above form facetgrid with some attributes
graph.map(plt.hist, 'total_bill', bins = 15, color ='orange')
# show the object
plt.show()
```

```python
tip.head()
```

```python
# Form a facetgrid using columns with a hue
graph = sns.FacetGrid(tip, col ='sex', hue ='smoker')
# map the above form facetgrid with some attributes
graph.map(sns.regplot, "tip", "total_bill").add_legend()
# show the object
plt.show()
```

```python
# importing packages
import seaborn
import matplotlib.pyplot as plt

# loading dataset
df = seaborn.load_dataset('tips')

# PairGrid object with hue
graph = seaborn.PairGrid(df, hue ='day')
# type of graph for diagonal
graph = graph.map_diag(plt.hist)
# type of graph for non-diagonal
graph = graph.map_offdiag(plt.scatter)
# to add legends
graph = graph.add_legend()
# to show
plt.show()
# This code is contributed by Deepanshu Rusatgi.
```

```python
# importing packages
import seaborn
import matplotlib.pyplot as plt

# loading dataset
df = seaborn.load_dataset('tips')

# PairGrid object with hue
graph = seaborn.PairGrid(df)
# type of graph for non-diagonal(upper part)
graph = graph.map_upper(sns.scatterplot)
# type of graph for non-diagonal(lower part)
graph = graph.map_lower(sns.kdeplot)
# type of graph for diagonal
graph = graph.map_diag(sns.kdeplot, lw = 2)
# to show
plt.show()
# This code is contributed by Deepanshu Rusatgi.
```
